```python
# uniform cost function
# returns the minimum cost (Bonus: returns path to minimum cost as well)
def  uniform_cost_search(start, goal, graph, cost):
    answer = {node: {'cost': float('inf'), 'path': []} for node in goal}
#     **************************************
#     ******** Write your code here ********
#     **************************************
    
    # Has node been visited?
    distanceFound = [False] * 30
    
    # Shortest total distance from start to node
    distanceFromStart = [None] * 30
    
    # Set distance from start at start to 0
    distanceFromStart[start] = 0
    
    # Version of initial graph with only the shortests paths from the specified starting node
    shortestGraph = [[] for i in range(30)]
    
    # Queue of nodes that are adjacent to nodes already visited
    # Each element has the form: [total distance from start, adjacent node already visited, the node itself]
    queue = []
    
    # Handle case of start in goal
    if (start in goal):
        answer[start]['cost'] = 0
    
    # Start the queue by getting the nodes adjacent to the start and sort them by distance
    for edge in graph[start]:
#         print(edge)
#         print(cost[start, edge])
        queue.append([cost[start, edge], start, edge])
    queue = sorted(queue)
#     print(queue)

    while (len(queue) > 0):

        # Find the cheapest node
        q = queue[0]
        del queue[0]
        
        # Don't visit node if already found
        if (distanceFound[q[2]] == False):
            
            # Visit node by updating data structures
            distanceFound[q[2]] = True
            distanceFromStart[q[2]] = q[0]
            shortestGraph[q[1]].append(q[2])

        #     print(distanceFound)
        #     print(distanceFromStart)
        #     print(shortestGraph)

            # Track costs for goal nodes
            if (q[2] in goal):
                answer[q[2]]['cost'] = q[0]
        
            # Update queue with the nodes neighboring the visited node
            for edge in graph[q[2]]:
                queue.append([q[0] + cost[q[2], edge], q[2], edge])

            # Sort queue by distance
            queue = sorted(queue)
#             print(queue)
            
    
    # Get paths
    for node in goal:
        
        # Handle unreachable goal
        if answer[node]['cost'] != float('inf'):
            
            # Calculate paths
            answer[node]['path'].insert(0, node)
            child = node
            while (child != start):
                for i in range(30):
                    if child in shortestGraph[i]:
                        answer[node]['path'].insert(0, i)
                        child = i
                        break
            
#     print(shortestGraph)
#     print(answer)
    return answer
 
# main function
if __name__ == '__main__':
     
    # create a graph with no more than 30 nodes
    graph, cost = [[] for i in range(30)], {}
 
    # add edges to the graph
    graph[0].append(4)
    graph[0].append(5)
    graph[0].append(16)
    graph[2].append(1)
    graph[3].append(1)
    graph[4].append(2)
    graph[4].append(3)
    graph[4].append(5)
    graph[5].append(8)
    graph[5].append(18)
    graph[6].append(3)
    graph[6].append(7)
    graph[8].append(16)
    graph[8].append(17)
    graph[16].append(17)
    graph[18].append(6)
    
 
    # add cost to each edge
    cost[(0, 4)] = 3
    cost[(0, 5)] = 9
    cost[(0, 16)] = 1
    cost[(2, 1)] = 2
    cost[(3, 1)] = 2
    cost[(4, 2)] = 1
    cost[(4, 3)] = 8
    cost[(4, 5)] = 2
    cost[(5, 8)] = 8
    cost[(5, 18)] = 2
    cost[(6, 3)] = 3
    cost[(6, 7)] = 2
    cost[(8, 16)] = 4
    cost[(8, 17)] = 4
    cost[(16, 17)] = 15
    cost[(18, 6)] = 1
    
    # set start state 
    start = 0
    
    # set goal state, there can be multiple goal states
    goal = [7]
    
    # call uniform_search_cost function to get the minimum cost to reach gaol. Bonus points for path
    # ****** You have to implement this function *****
    min_cost_info = uniform_cost_search(start, goal, graph, cost)

    for node, info in min_cost_info.items():
        print(f'Minimum cost from {start} to {node} is {info["cost"]}')
#         print(graph)
#         print('--------')
#         print(cost)
      
    #   **** Bonus ****
        print(f'Path: {info["path"]}')
```
